```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Problem 1
```

```python
#1a

samples = pd.read_csv('SpotifyFeatures.csv')    #load data
n_samples, n_features = samples.shape    #returns tuple from data
print("Number of samples:", n_samples)
print("Number of features:", n_features)

# 1b

#filter pop and classical music by labeling them
pop_classic = samples.copy()
rows_to_keep = []

for index, row in pop_classic.iterrows():    #check if it is pop or classical
    genre_value = row['genre'] 
    
    if genre_value == 'Pop' or genre_value == 'Classical':
        rows_to_keep.append(row)  

pop_classic = pd.DataFrame(rows_to_keep)


pop_classic['label'] = pop_classic['genre'].apply(lambda x: 1 if x == 'Pop' else 0)    # adds new clomumn where pop = 1 and classic = 0


pop_count = pop_classic[pop_classic['genre'] == 'Pop'].shape[0]
classic_count = pop_classic[pop_classic['genre'] == 'Classical'].shape[0]

print("Number of 'Pop' samples:", pop_count)
print("Number of 'Classical' samples:", classic_count)

# 1c

X = pop_classic[['liveness', 'loudness']].values
y = pop_classic['label'].values


pop_indices = np.where(y == 1)[0]
classical_indices = np.where(y == 0)[0]


# shuffle the samples
np.random.seed(42)
np.random.shuffle(pop_indices)    
np.random.shuffle(classical_indices)


# split the samples into 80% training data and 20% testing data
pop_split_index = int(0.8 * len(pop_indices))
classical_split_index = int(0.8 * len(classical_indices))

pop_train_indices = pop_indices[:pop_split_index]
pop_test_indices = pop_indices[pop_split_index:]

classical_train_indices = classical_indices[:classical_split_index]
classical_test_indices = classical_indices[classical_split_index:]


# combine training indices of pop and classical aswell as for the test indices
train_indices = np.concatenate([pop_train_indices, classical_train_indices])
test_indices = np.concatenate([pop_test_indices, classical_test_indices])


np.random.shuffle(train_indices)
np.random.shuffle(test_indices)


X_train = X[train_indices]
y_train = y[train_indices]
X_test = X[test_indices]
y_test = y[test_indices]


print("Training set size:", X_train.shape[0], "samples")
print("Test set size:", X_test.shape[0], "samples")


print("Class distribution in training set:", sum(y_train == 1), "Pop", sum(y_train == 0), "Classical")
print("Class distribution in test set:", sum(y_test == 1), "Pop", sum(y_test == 0), "Classical")

# 1d

# create a figure and scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='yellow', label='Pop')
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='green', label='Classical')

plt.xlabel('Liveness')
plt.ylabel('Loudness')
plt.title('Liveness vs Loudness')
plt.legend()
plt.show()
```

```python
#2a
# sigmoid function to convert the input to a value between 0 and 1
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# this function calculates how far the predictions are from the the actual label
def compute_loss(y_true, y_pred):
    epsilon = 1e-10  
    return -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))

class LogisticRegression:
    def __init__(self, learning_rate, epochs, batch_size):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.weights = None
        self.bias = None

    # set weights and bias, creates empty list to track the loss
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        losses = []

        # training loop
        for epoch in range(self.epochs):
            indices = np.arange(n_samples)
            np.random.shuffle(indices)  
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            
            epoch_loss = 0

            # mini batch gradient descent
            for i in range(0, n_samples, self.batch_size):
                X_batch = X_shuffled[i:i + self.batch_size]
                y_batch = y_shuffled[i:i + self.batch_size]
                
                z = np.dot(X_batch, self.weights) + self.bias
                y_pred = sigmoid(z)
                
                error = y_pred - y_batch
                dw = np.dot(X_batch.T, error) / len(y_batch)
                db = np.mean(error)
                
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db
                
                epoch_loss += compute_loss(y_batch, y_pred)
            
            epoch_loss /= (n_samples // self.batch_size)
            losses.append(epoch_loss)

            # print the loss every 50 epochs
            if epoch % 50 == 0:
                print(f"Epoch {epoch+1}/{self.epochs}, Loss: {epoch_loss:.4f}")

        # plot the loss over the epochs
        plt.plot(range(self.epochs), losses)
        plt.title("Training Error")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.show()

    # calculate predicted probabilities via the sigmoid function
    def predict_prob(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return sigmoid(linear_model)
    
    def predict(self, X):
        y_pred = self.predict_prob(X)
        return [1 if i > 0.5 else 0 for i in y_pred]

X = pop_classic[['liveness', 'loudness']].values
y = pop_classic['label'].values

X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

lr = LogisticRegression(learning_rate=0.001, epochs=500, batch_size=32)

lr.fit(X_train, y_train)

y_train_pred = lr.predict(X_train)
train_accuracy = np.mean(y_train_pred == y_train)
print(f"Training accuracy: {train_accuracy * 100:.2f}%")

y_test_pred = lr.predict(X_test)
test_accuracy = np.mean(y_test_pred == y_test)
print(f"Test accuracy: {test_accuracy * 100:.2f}%")

# 2b
# check accuracy on the test set
accuracy_diff = train_accuracy - test_accuracy
print(f"Difference between training and test accuracy: {accuracy_diff * 100:.2f}%")

if accuracy_diff > 0.05:
    print("There is a significant difference between training and test accuracy, this may be due to overfitting.")
else:
    print("The difference between training and test accuracy is small, this means it generalizes well.")

print(f"The test accuracy score is {test_accuracy * 100:.2f}%.")

# 2c
plt.figure(figsize=(10, 6))

# plot the training data points
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='yellow', label='Pop')
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='green', label='Classical')

x_values = np.linspace(X_train[:, 0].min(), X_train[:, 0].max(), 100)
decision_boundary = -(lr.weights[0] / lr.weights[1]) * x_values - (lr.bias / lr.weights[1])

# plot decision boundary
plt.plot(x_values, decision_boundary, color='black', label='Decision Boundary')

plt.xlabel('Liveness')
plt.ylabel('Loudness')
plt.title('Liveness vs Loudness with Decision Boundary')
plt.legend()
plt.show()
```

```python
def confusion_matrix(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # calculate true positives, true negatives, false positives, and false negatives
    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    
    # return confusion matrix 
    return pd.DataFrame({
        'Predicted 0 (Classical)': [TN, FN],
        'Predicted 1 (Pop)': [FP, TP]
    }, index=['Actual 0 (Classical)', 'Actual 1 (Pop)'])

# generate confusion matrix from test set
conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
